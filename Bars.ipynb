```python
import pandas as pd
```

```python
boxoffice = pd.read_csv("https://raw.githubusercontent.com/rhuanca/sample-data-sets/main/boxoffice.csv")
boxoffice['count'] = 1
boxoffice
```

```python
boxoffice.dtypes
```

```python
boxoffice[['studio', 'count']] \
  .groupby(by='studio').sum() \
  .sort_values(by='count', ascending=False)
```

```python
boxoffice[['studio', 'count']] \
  .groupby(by='studio').sum() \
  .sort_values(by='count', ascending=False) \
  .plot(kind='bar', figsize=(20, 8))
```

```python
boxoffice[['studio', 'count']] \
  .groupby(by='studio').sum() \
  .sort_values(by='count', ascending=False) \
  .head(10) \
  .plot(kind='bar', figsize=(20, 10))
```

```python
boxoffice[['studio', 'count']] \
  .groupby(by='studio').sum() \
  .sort_values(by='count', ascending=False) \
  .head(10) \
  .plot(kind='bar', figsize=(20, 10), rot=0)
```

```python
boxoffice[['studio', 'count']] \
  .groupby(by='studio').sum() \
  .sort_values(by='count', ascending=False) \
  .head(10) \
  .sort_values(by='count', ascending=True) \
  .plot(kind='barh', figsize=(20, 10), rot=0)
```

```python
boxoffice[['studio', 'count']] \
  .groupby(by='studio').sum() \
  .sort_values(by='count', ascending=True) \
  .tail(10) \
  .plot(kind='barh', figsize=(20, 10), rot=0)
```
