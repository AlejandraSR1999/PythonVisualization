```python
import pandas as pd
import plotly.express as px
```

```python
raw = pd.read_csv("https://raw.githubusercontent.com/rhuanca/sample-data-sets/main/API_IT.NET.USER.ZS_DS2_en_csv_v2_4325432.csv")
raw
```

```python
raw['Indicator Name'].unique()
```

```python
countries = ['BOL', 'PER', 'ARG', 'BRA', 'COL', 'CHL', 'MEX']
```

```python
internet = raw[raw['Country Code'].isin(countries)]
# internet = raw.copy()
internet = internet[['Country Code', '2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', 
             '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020']]


internet
```

```python
internet[internet["Country Code"] == 'ARG']
```

```python
data = internet[internet["Country Code"] == 'ARG'] 
data = data.set_index('Country Code')

data.plot(kind='bar', figsize=(20,8), rot=0)
```

```python
data = internet.copy()
data = data.set_index('Country Code')

data.plot(kind='bar', figsize=(30,8), rot=0)
```

```python
data = internet.set_index('Country Code')
data = data.sort_values(by='2020', ascending=False)
fig = px.imshow(data)
fig
```

```python
data = internet.set_index('Country Code')
data = data.sort_values(by='2000', ascending=False)
fig = px.imshow(data)
fig
```
